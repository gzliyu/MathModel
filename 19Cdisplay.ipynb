```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
import csv
```

```python
f = open("new_location.csv","w",newline="")
w = csv.writer(f)
loc = pd.read_csv("location.csv")
for row in loc.iterrows():
    name = row[1][0]
    w.writerow([name[0:-3],name[-2:],row[1][1],row[1][2]])      
    
    
```

```python
nLoc = pd.read_csv("new_location.csv")
data = pd.read_csv("data.csv")
data=data[data["YYYY"]==2010]

colorDict=['#1f77b4','#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
state=["KY","OH","PA","VA","WV"]
y=2014
i=0
for s in state:
    d=data[data["State"]==s]
    c=[]
    x=[]
    y=[]
    size=[]
    location=nLoc[nLoc["state"]==s]

    for row in d.iterrows():
        county=row[1]["COUNTY"]
        
        if county not in c:
            size.append(row[1]["TotalDrugReportsCounty"]/1000)

            c.append(county)
            l=location[location["county"]==county]
            # print(l["lat"].values[0])
            # print(l["lat"].values[0])
            try:
                x.append(l["lat"].values[0])
                y.append(float(l["lot"].values[0]))
            except IndexError:
                continue


    plt.scatter(x=x,y=y,c=colorDict[i],s=size)
    i+=1
plt.show()
    
    







```
